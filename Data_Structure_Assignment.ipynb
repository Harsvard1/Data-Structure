1. Mutable Data Types
Mutable data types can be changed after they are created. You can modify their content (add, remove, or update elements) without creating a new object.
my_list = [1, 2, 3]
my_list.append(4)  # Adds 4 to the list
print(my_list)  # Output: [1, 2, 3, 4]

2. Immutable Data Types
Immutable data types cannot be changed after they are created. Any modification creates a new object instead of altering the existing one.
my_string = "Hello"
my_string = my_string + " World"  # Creates a new string
print(my_string)  # Output: "Hello World"

A hash table is a data structure that stores key-value pairs, allowing for efficient data retrieval. It uses a hash function to map keys to specific locations in memory (called buckets or slots) where their corresponding values are stored. Hash tables are highly efficient for operations like searching, insertion, and deletion, typically providing constant-time access.

In Python, the dictionary (dict) is implemented using a hash table. This means that Python dictionaries leverage the hash table's efficiency to store and retrieve key-value pairs quickly.

In Python, sets automatically remove duplicate values. A set is an unordered collection of unique elements, meaning that it will only store one occurrence of each element. If you try to add a duplicate element to a set, it simply won't be added again.

How Sets Handle Duplicates
When you add an element to a set, Python checks if the element already exists in the set. If the element is already present, it won't be added again.
The set data structure is based on hash tables, which means it checks the hash value of an element for uniqueness before adding it.

Ans. 14

The in keyword in Python is used to check for membership, but it works differently for lists and dictionaries because the way data is stored and accessed in each of these structures differs.

1. in for Lists:
When using the in keyword with a list, it checks whether a value exists in the list. It does not check for the index or position of the element, but rather whether the element is present as a value in the list.

2. in for Dictionaries:
When using the in keyword with a dictionary, it checks for the keys of the dictionary, not the values. If the key exists in the dictionary, it returns True; otherwise, it returns False.

In Python, dictionaries are implemented using hash tables, which allow for very efficient access to their elements. The time complexity of accessing elements in a dictionary is typically O(1), which means it takes constant time, regardless of the size of the dictionary.

The average case time complexity for accessing elements in a dictionary (whether by key lookup, adding, or deleting) is O(1), which is highly efficient.
In rare cases, the time complexity can degrade to O(n) due to hash collisions, but this is unusual because Python's hash table implementation is designed to handle collisions efficiently.
